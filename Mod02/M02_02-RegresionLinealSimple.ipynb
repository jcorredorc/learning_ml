# Regresión lineal simple por el método de los mínimos cuadrados
La regresión lineal simple se caracteriza por tener una sola variable predictora $x$ en el modelo desarrollado, con un termino independiente constante $\theta_{0}$ (intercepción de la recta con el eje $y$), y un peso $\theta_{1}$ asignado a la variable $x$ que es la pendiente de la recta de regresión. Estos valores de $theta$ son los que va a predecir el modelo de regresión simple.

$$\large{y=\theta_{0}+\theta_{1}x+\epsilon}$$

El modelo matricial general es el siguiente para el modelo general de regresión lineal, en el caso de la regresión lineal simple solo existiría una columna para la variable predictora $x$ y solo dos valores de los parámetros $\theta$ $(\theta_{0}$ y $\theta_{1})$.

$Y=\large{
\left[
\begin{array}{}
y^{(1)}\\
y^{(2)}\\
y^{(3)}\\
y^{(3)}\\
:\\
y^{(m)}
\end{array}
\right]
}, \large{
\ \ \ X=
\left[
\begin{array}{}
1&x_{1}^1&x_{2}^1&x_{3}^1&...&x_{n}^1\\
1&x_{1}^2&x_{2}^2&x_{3}^2&...&x_{n}^2\\
1&x_{1}^3&x_{2}^3&x_{3}^3&...&x_{n}^3\\
1&x_{1}^4&x_{2}^4&x_{3}^4&...&x_{n}^3\\
:&:&:&:&:&:\\
1&x_{1}^m&x_{2}^m&x_{3}^m&...&x_{n}^m\\
\end{array}
\right]
}, \ \ \  \large{
\theta=
\left[
\begin{array}{}
\theta_{0}\\
\theta_{1}\\
\theta_{2}\\
\theta_{3}\\
:\\
\theta_{n}
\end{array}
\right]
}$

### Revisar correlación entre las variables `precio` y `num_hab`

```python
# Ver la correlación entre las variables del modelo lineal simple
df[['precio', 'num_hab']].corr()
## El 69.63% del valor real del precio se pueden ver identificados por el num_hab
## 1-correlacion = 0.30, me explica que el 30% de las varianzas de los precios de las propiedades
```

```python
from scipy.stats import pearsonr

# test_cor devuelve un vector con dos valores [0]-[1]
test_cor = pearsonr(x = df['num_hab'], y = df['precio'])
print('Coeficiente de correlación: ', test_cor[0])
print('P-valor: ', test_cor[1])
```

Por cualquiera de los dos métodos se puede contrastar el valor de la correlación de Pearson, sin embargo, si se quiere ver la significancia del contraste de la hipótesis nula de la prueba, no relación entre las variables, se tiene una significancia (p-valor) por debajo de 0.05 lo que indica que se debe rechazar la hipótesis nula ya que hay evidencia estadística que nos permite inferir que las variables están estrechamente relacionadas.

### Modelo de regresión lineal simple

```python
# Importar modulo que desarrolla la regresión lineal por el método de los mínimos cuadrados
# mostrando sus estadísticos
import statsmodels.api as sn
```

```python
# Recordar que para efectos del calculo del modelo, el valor del intercepto esta acompañado por una
# variable x que se asume con un valor de 1.
X = sn.add_constant(df['num_hab']) # Generar constantes iniciales de los parámetros
```

```python
X
```

```python
# Crear el objeto del modelo y realizar el ajuste (fit) de los parámetros
linear_model = sn.OLS(df['precio'], X).fit()  # minimos cuadrados
```

```python
# Ver los estadísticos del modelo y las significancias de las pruebas
linear_model.summary()
```

---
El valor de la significancia del estadístico F, indica que nuestra única variable predictora `num_hab` puede explicar la variación de los precios de las propiedades del conjunto de datos.

R-squared: correlacion media en 0.48 (un valor ideal sería 0.7 o 0.8).

El modelo sería  y= -34.6592 + 9.0997 X

```python
# Visualizar la nube de puntos
sns.scatterplot(x = df['num_hab'], y = df['precio'])
```

```python
# Opción 1: Visualizar la línea de regresión generada por el modelo
sns.lineplot(x = df['num_hab'], y = -34.6592 + 9.0997*df['num_hab'])
```

```python
# Opción 2: Visualizar nube de puntos y línea de regresión # usando la función regplot
sns.regplot(x = df['num_hab'], y = df['precio'], ci =90)
# patch es el valor de confianza ci =90
```

```python
# Opción 3: Visualizar la nube de puntos y línea de regresión con la función jointplot (kind = 'reg')
sns.jointplot(x = df['num_hab'], y = df['precio'], kind = 'reg', color = 'blue')
```

```python
# Con la función jointplot ( parámetro kind = 'resid') se pueden visualizar los residuos (alturas) de
# los valores observados vs los predichos
# Residuos son las distancias del valor real al predicho, error
sns.jointplot(x = df['num_hab'], y = df['precio'], kind = 'resid', color = 'darkorange')
```



* ¿Qué tipo de comportamiento tienen estos errores?


* ¿Sería el modelo líneal el que mejor representa estos datos para desarrollar predicciones a partir de ellos?

```python
# Al desarrollar una función con la ecuación de la linea de regresión se pueden predecir los valores
# de los precios dada el número de habitaciones
def prediccion(x):
    y = -34.6592 + 9.0997 * x
    print('El valor proyectado para', x, 'habitaciones es', '{0:.2f}'.format(y))
    
prediccion(10)
```

# Regresión lineal multiple
### Por el método de los mínimos cuadrados

```python
# Crear un nuevo dataframe sin la columna precio (axis=1 -> borra columna), vector de caracteristicas
X_multiple = df.drop('precio', axis = 1) ## 
```

```python
# Verificar la matriz de atributos
X_multiple.head()
```

```python
y_multiple = df['precio'] ## def variable a predecir
```

```python
y_multiple
```

```python
X_multiple_constante = sn.add_constant(X_multiple) ## agregar la constante
```

```python
X_multiple_constante
```

```python
lin_model_mult = sn.OLS(y_multiple, X_multiple_constante).fit() ## minimos cuadrados
```

```python
lin_model_mult.summary()
## la columna de coef del summary son los thetas del modelo
```

### Por el módulo sklearn (utilizando gradiente de descenso)

```python
# Importar el modulo y la función necesaria pare el modelo de regresión lineal
from sklearn.linear_model import LinearRegression
```

```python
# Crear el modelo
lin_model_sk = LinearRegression()
```

```python
# Ajustar el modelo a los datos
lin_model_sk.fit(X_multiple, y_multiple)
```

```python
# Ver los valores del intercepto y de los coeficientes
inter = lin_model_sk.intercept_
coef = []
coef = lin_model_sk.coef_
print('Intercepto theta 0:','{0:.3f}'.format(inter))

# Mostrar los coeficientes de las variables regresoras
theta = 1
for i in coef:
    print('theta', theta, ':', '{0:.3f}'.format(i))
    theta += 1
```
