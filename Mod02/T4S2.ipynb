## Data Science
La ciencia de datos o Data Science es una interdisciplina que busca aplicar los metodos matematicos y computacionales al analisis de los datos (especialmente en grandes volumenes) de tal manera que pueda extraerse mejor informacion de estos. Asi, un cientifico de datos debe tener habilidades en matematicas, estadistica y computacion, ademas del conocimiento experto en el area de donde provienen los datos, que le permita no solo analizar los datos sino ademas tomar decisiones relevantes en base a estos. Una gran ventaja es que muchas de las tecnicas matematicas que pueden ayudar en el manejo de los datos estan implementadas en varias librerias de Python.

## La libreria NumPy
Numpy es un paquete fundamental para computacion cientifica, que puede fuencionar como contenedor multidimensional de datos genericos con bastante eficiencia. Contiene muchas herramientas del algebra lineal, transformadas, generadores de numeros aleatorios, herramientas para integrar C,C++, Fortran, entre otras cosas. Las herramientas de NumPy son por lo general bastante rapidas, pues estan basadas en las librerias de lenguaje C. 

### Los arreglos de NumPy

La funcionalidad mas comun de NumPy son los arreglos. De manera sencilla, los arreglos son la representacion de las matrices del algebra lineal.

```python
# Para llamar a la libreria se usa import
# Se puede agregar un atajo para llamarla mas adelante
import numpy as np
```

```python
# Se puede construir un arreglo a partir de una lista
l1=[1,2,3]
# Para construir un arreglo se usa la funcion array
A=np.array(l1)
A
```

```python
# La lista para construir el arreglo se puede escribir dentro 
# del llamado de la funcion
B=np.array([[2,3],[4,5]])
B
```

Cada lista que se pasa dentro de los parametros se toma como una fila de la matriz. Tambien se pueden construir arreglos de manera secuencial, usando la funcion arange()

```python
# La funcion arange tiene tres parametros: un valor de inicio,
# un valor limite y un tamaño de paso. El valor limite se omite
np.arange(2,100,2)
```

Tambien se pueden construir arreglos de ceros y unos

```python
# la funcion zeros pide un argumento de forma para el 
# tamaño del arreglo, que se pasa como una tupla
print(np.zeros((2,3,4)))
```

```python
# Funciona igual para la funcion ones()
np.ones((4,5))
```

Tambien existe la funcion linspace(), que genera un vector con una particion entre dos puntos, con un numero fijo de puntos, o la funcion eye() que crea una matriz identidad

```python
# los argumentos de linspace son valor inicial, valor final, numero de puntos
np.linspace(3,8,10)
```

```python
# eye solo lleva un argumento, pues las matrices identidad son cuadradas
np.eye(5)
```

Tambien se pueden generar matrices de numeros aleatorios, con las funciones rand(), randn(), randint(). estas estan dentro de la sublibreria random().

```python
# La funcion rand genera numeros aleatorios entre 0 y 1
print(np.random.rand(5,2))
# randn genera numeros aleatorios basados en una normal estandar.
print(np.random.randn(2,3))
# randint genera numeros enteros de manera aleatoria con tres parametros,
# valor inicial, valor final, numero de elementos. El valor final
# se omite.
print(np.random.randint(-2,2,(2,3)))
```

Los arreglos tiene varios atributos, que son pequeñas funciones preconstruidas con el arreglo. 

```python
# shape devuelve el tamaño del arreglo
A=np.array([[2,3,1,4,5],[5,1,-2,3,3]])
A.shape
```

```python
# reshape() reacomoda el arreglo a una nueva forma. 
# Es necesario que tenga el mismo numero de elementos.
print(A)
print(A.reshape((5,2)))
```

```python
# Para buscar un valor maximo o minimo
print(A.max(),A.min())
# Para buscar los indices de dichos valores
print(A.argmax(),A.argmin())
```

Es posible buscar elementos particulares de un arreglo usando los indices de posicion. Para llamar un indice se usan []

```python
# Para llamar a la primera fila del arreglo
print(A[0])
# Para llamar al segundo elemento en la primera fila
print(A[0][1])
```

Es posible tambien modificar valores especificos del arreglo, o todo un rango de indices

```python
print(A)
# Cambiando un valor del arreglo
A[0][1]=-5
print(A)
# Cambiando toda la primera fila 
A[0]=0
print(A)
# Cambiando todos los elementos de la primera fila,
# menos el primero
A[0][1:]=-3
print(A)
```

Es importante anotar que salvo que se cree de manera explicita, no se generan copias de las variables en Python, solo atajos. Esto permite ahorrar memoria.

```python
A=np.array([1,2,3])
print(A)
B=A
B[0]=0
print(A)
```

Se pueden efectuar operaciones sobre todo el arreglo: las algebraicas son matriciales, pero tambien se pueden efectuar operaciones booleanas y de operadores.

```python
print(A==0)
print(A[A%2==0])
```

```python
# Las operaciones algebraicas son similares a las operaciones con matrices
# que se realizan componente a componente.
A=np.random.randint(-10,10,(2,3))
print(A)
print(A+A)
print(2*A)
print(A*A)
# Las funciones universales dentro de NumPy tambien se pueden aplicar a arreglos
print(np.sin(A))
```

```python
# Para transponer el arreglo, transpose()
# Para multiplicar matrices, matmul()
np.matmul(A,np.transpose(A))
```

```python
# El arreglo es un objeto iterable. 
# Por defecto se itera por filas
for i in A:
  print(i)
```

```python
# Es posible "aplanar" el arreglo
for i in A.flat:
  print(i)
```

Aunque quiza la operacion que mas se recuerda cuando se habla de matrices son los sistemas de ecuaciones y las operaciones del algebra lineal. NumPy tiene una libreria dedica al algebra lineal, linalg. Dentro de esta se puede usar solve() para solucionar sistemas de ecuaciones, siempre que la matriz de coeficientes sea cuadrada e invertible.

```python
A=np.random.randint(-3,3,(2,2))
b=np.random.randint(-3,3,(2,1))
np.linalg.solve(A,b)
```

```python
np.mean(A)
```

## SymPy
El paquete Sympy es un paquete de calculo de algebra simbolica. Permite realizar operaciones con objetos matematicos, como factorizaciones, limites, derivadas, integrales, etc. 

```python
# Si se calcula una raiz cuadrada con el paquete math
import math
math.sqrt(8)
```

```python
# La raiz cuadrada de 8 es un numero irracional, asi que al 
# pedir una raiz cuadrada se obtiene una aproximacion. 
# Si se ejecuta el mismo calculo en SymPy
import sympy
sympy.sqrt(8)
```

SymPy puede trabajar con variables de manera abstracta como se hace en matematicas. Se definen como objetos simbolos

```python
from sympy import Symbol
x=Symbol('x')
y=Symbol('y')
pol=x**2-2*x*y+y**2
pol
```

```python
from sympy import factor, expand
factor(pol)
```

Con los simbolos se pueden realizar muchos procesos del calculo de manera automatica

```python
from sympy import *
# Una derivada
diff(sin(x**2))
```

```python
# una integral
integrate((x)/(x**2+1),x)
```

```python
# un limite
limit((x**2-1)/(x-1),x,1)
```

```python
# buscar raices de polinomios
solve(x**2-2*x-5,x)
```

La documentacion de SymPy es extensa, y abarca procesos en muchas areas, desde el calculo a la teoria de categorias, la combinatoria, ecuaciones diferenciales, geometria, fisica, etc. Es una muestra de lo versatil que puede llegar a ser un lenguaje de programacion

## Pandas
Una de las librerias mas populares para trabajar con datos, Pandas permite realizar analisis de datos de forma eficiente, ejecutar tareas de limpieza y preprocesamiento, visualizaciones, etc. Esta construida sobre NumPy y en algunos casos se considera una version muy potente de Excel, aunque Pandas posee muchas mas caracteristicas. 

Pandas puede definir dos objetos basicos: las series y los DataFrames. Una serie es similar a un arreglo de NumPy donde ademas de los indices se pueden usar etiquetas para determinar la posicion de un objeto sobre la serie. Puede ademas contener cualquier objeto de Python.

```python
# Para importar Pandas
import pandas as pd
```

```python
# Para crear una serie, puede usarse la funcion Series
s=pd.Series(data=[2,3,4])
s
```

Sin un conjunto de etiquetas, se asignan etiquetas numericas similares a una indexacion usual.

```python
s=pd.Series(data=[2,3,4],index=['a','b','c'])
s
```

```python
s['a']
```

Pueden crearse series a partir de arreglos o diccionarios, tambien. La serie puede almacenar cualquier dato, incluso funciones.

```python
s=pd.Series(data=[2,len,4])
s
```

A partir de las series es posible revisar los DataFrames. Es posible pensar en un DataFrame como un conjunto de series (columnas) que comparten los mismos indices. Los DataFrame estan inspirados en un objeto del lenguaje R y son la herramienta mas comun en Python para trabajar con bases de datos.

```python
# Para llamar la libreria
import pandas as pd
```

```python
# Un dataframe requiere al menos de un arreglo como parametro. 
# Pueden agregarse indices y nombres de columnas, pues el dataframe 
# tiene una estructura similar a una matriz.
df=pd.DataFrame(np.random.randn(5,5),index=['A','B','C','D','E'],columns=['col1','col2','col3','col4','col5'])
```

```python
df
```

```python
# Se puede hacer un llamado a una columna usando su nombre
df['col1']
```

```python
type(df['col1'])
```

```python
# Se pueden construir columnas nuevas a partir de columnas ya existentes o de datos nuevos
df['col6']=df['col1']**2
df
```

```python
# Tambien es posible quitar columnas del arreglo
df.drop('col6',axis=1,inplace=True) 
# axis indica si se trabaja en filas (0) o columnas (1)
# inplace indica si el cambio modifica o no el dataframe
df
```

```python
# Las filas tambien pueden ser llamadas,
# con el comando loc si se usa la etiqueta del indice
# o con el comando iloc si se usa la posicion.
print(df.loc['A'])
print(df.iloc[0])
```

```python
# selecccionar varias columnas
df[['col1','col3']]
```

Se pueden realizar operaciones sobre las columnas o filas del dataframe, algunas propias de Pandas.

```python
# Para calcular el promedio de una columna
df['col1'].mean()
```

Esta funcion estadistica tambien se puede obtener, junto a otros indicadores utiles, por medio del metodo describe

```python
df.describe()
```

Al igual que NumPy, se pueden usar condicionales logicos para buscar datos.

```python
df>0
```

```python
df[df>0]
```

Los registros faltantes se notan como NaN ('not a number'). Generalmente esos valores se retiran o se reemplazan.

```python
# dropna() quita por defecto las filas con valores NaN. Se puede cambiar
# a columnas con un axis=1 y hacer el cambio permanente con un inplace=True
df[df>0].dropna()
```

```python
# Para cambiar un registro NaN por un valor, se usa fillna
df[df>0].fillna(value=0)
```

Tambien se pueden buscar registros que cumplan diversas condiciones sobre las columnas

```python
df[(df['col1']>1) & (df['col5']<0)]
```

Pandas tambien soporta multi-indices, lo que permite guardar muchas series de datos en un mismo dataframe

```python
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside))
hier_index = pd.MultiIndex.from_tuples(hier_index)
df = pd.DataFrame(np.random.randn(6,2),index=hier_index,columns=['A','B'])
df
```

Tambien se pueden realizar agrupaciones de datos basados en valores. Para esto se usa la funcion groupby.

```python
# Se puede usar un diccionario para crear un dataframe
df=pd.DataFrame({'Sede':['Bogota','Bogota','Cali','Cali','Medellin','Medellin'],
                 'Alumno':['Juan','Andrea','Andrea','Carlos','Mario','Luis'],
                 'Edad':[18,21,25,17,20,21]})
df
```

```python
# Agrupemos por sede 
df.groupby('Sede')
```

La funcion groupby genera un objeto de pandas. Para sacar informacion se debe usar una funcion de agregacion (una funcion numerica que reduzca los datos de acuerdo a algun parametro)

```python
df.groupby('Sede').mean()
```

Notese que esta funcion de agregacion solo se ejecuta sobre la columna que tiene variables numericas. Distintas funciones generan diferentes resultados.

```python
df.groupby('Sede').min()
```

Otra forma util de sacar informacion sobre el DataFrame son atributos

```python
# informacion general del tipo de variables
df.info()
```

```python
# los nombres de las columnas
df.columns
```

```python
# valores unicos de una columna
df['Sede'].unique()
```

```python
# Conteo de frecuencias
df['Sede'].value_counts()
```

Si se quiere aplicar alguna transformacion a los datos, se puede usar una funcion apply en composicion con otra funcion, o una expresion lambda

```python
df['Edad'].apply(lambda x: x**2)
```

Se puede ordenar un dataframe de acuerdo a los valores de una columna

```python
df.sort_values('Edad')
```

Tambien pueden combinarse dataframe, usando principalmente tres metodos
* concat, que pega dos dataframes uno tras otro
* merge, que combina los dataframe a partir de valores comunes en las columnas
* join, que combina los dataframe a partir de valores comunes en los indices.

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7])
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
# Concatenar
pd.concat([df1,df2,df3])
```

```python
izq = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
der = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']}) 
```

```python
# Fusionar 
pd.merge(izq,der,on='key')
```

merge tiene un argumento extra, how, que dice como se usan las claves de la fusion.

```python
izq = pd.DataFrame({'key': ['K0', 'K1', 'K0', 'K0'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
der = pd.DataFrame({'key': ['K0', 'K0', 'K2', 'K2'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']}) 
```

```python
pd.merge(izq,der,on='key',how='left')
```

De estas herramientas se puede ver la fuerza de Python para trabajar con bases de datos, con limpieza y recuperacion de informacion.
