## Sintaxis: 
La sintaxis es "(...) la parte de la gramática que estudia las reglas y principios que gobiernan la combinatoria de constituyentes sintácticos y la formación de unidades superiores a estos, como los sintagmas y las oraciones gramaticales". De manera mas simple, son las reglas que dicen el orden que deben seguir las palabras dentro de una oracion para que esta se considere bien escrita, y como se relaciona una palabra con las demas. Para el lenguaje de programacion, la sintaxis es el conjunto de reglas para escribir el codigo que permitan que el interprete las reconozca sin ambiguedades.

## Estructuras de flujo

Aparte de las variables, se pueden construir estructuras que ejecutan instrucciones basadas en condiciones, en iteraciones, o en cantidades fijas.

### Condicionales: if, elif, else:
Los flujos condicionales permiten ejecutar una serie de instrucciones siempre que una condicion booleana se cumple. La estructura basica de estos se basa en el condicional if (si):

```python
# La estructura del if
# if condicion=verdadera:
#    instruccion
```

Es importante anotar dos cosas: primero, la condicion definida en if debe cumplirse para poder entrar a ejecutar las instrucciones, que se escriben en la siguiente linea del if, y la condicion debe ir seguida de dos puntos. Segundo, dichas instrucciones conforman un bloque de codigo aparte, delimitado por la indentacion (espacios) con respecto a las lineas principales. El espacio que debe usarse para delimitar un bloque dentro de un if (y en general dentro de cualquier flujo) es una tabulacion.

```python
print('Aqui estamos fuera del condicional')
print('Se va a probar una condicion')
if 1<2:
  print('   Estamos dentro del if')
print('Python detecta automaticamente la salida del ciclo por la indentacion')
```

```python
var=input('Digite su edad: ')
var=float(var)
if var>30:
  print('La mitad de su edad es {}'.format(var/2))
  print('Usted está muy viejo')
print('FIN')
```

Dentro del if puede agregarse una serie extra de instrucciones que se ejecutan si la condicion del if no se cumple, usando un else:

```python
# La estructura del if-else
# if condicion=verdadera:
#    instruccion1
# else:
#    instruccion2
# Notese que if y else estan al mismo nivel de indentacion
```

```python
print('Un if-else')
var=int(input('Ingrese un digito '))
if var>5:
  print('El digito es mayor que 5')
else: #notese que else tambien tiene : al final de la linea
  print('El digito es menor o igual que 5')
```

Es posible añadir mas condiciones dentro de un ciclo if-else, usando la instruccion elif entre el if y el else. 

```python
print('Un if-elif-else')
var=int(input('Ingrese un digito '))
if var>7:
  print('El digito es mayor que 7')
elif var>5: #notese que elif lleva una condicion al igual que if
  print('El digito es mayor que 5 y menor o igual a 7')
elif var>3:
  print('El digito es mayor que 3 y menor o igual que 5')
else:
  print('El digito es menor o igual a 3')
```

En un in-elif-else, se verifica una a una las condiciones enunciadas en if y elif, en cuanto una se verifica se ejecutan las instrucciones del bloque correspondiente y se interrumpe el condicional: las instrucciones que queden son ignoradas.

### Ciclos: for, while, range:
Los ciclos permiten ejecutar una serie de instrucciones varias veces mientras una condicion se cumpla , como los ciclos while, o mientras se recorre un objeto iterable (un objeto con enumeracion, como una lista) con los ciclos for. 

```python
# Un ciclo while ejecuta una serie de instrucciones 
# mientras una condicion se cumpla.
# Por lo general, al interior del ciclo se puede modificar la variable 
# que se usa en la condicion.
i=1
while i<5:
  print('Vamos en la ejecucion '+str(i))
  i=i+1
```

```python
# un ciclo for pasa por cada uno de los elementos de un 
# objeto iterable, como una lista
lista=[1,2,3,4]
for i in lista:
  print('El elemento {} de la lista es {}' .format(i,lista[i-1]))
```

```python
# Se puede definir un objeto iterable usando la funcion range(), 
# dando como argumento el numero de iteraciones que queremos
for i in range(5):
  print(i)
```

### Break y continue
Es posible alterar el flujo de un ciclo usando dos comandos:
* *break* detiene la ejecucion de un ciclo y sale del mismo 
* *continue* detiene la ejecucion de la iteraccion actual del ciclo y salta a la siguiente iteracion.

### Listas por comprension
Se puede ejecutar una operacion sobre cada uno de los elementos de una lista y definir asi una lista nueva. Para esto se usa la notacion de las listas y el ciclo for

```python
# La definicion de la lista se hace con []
# dentro de estos va una instruccion para
# los elementos que se quieren iterar, seguida de 
# un for y el objeto iterable.
lista=[1,2,3,4,5]
[item**2 for item in lista]
```

Se pueden combinar condicionales dentro de la estructura

```python
# Para volver a revisar la lista buscando los numeros pares
[item for item in lista if item%2==0]
```

## Funciones
Ademas de las funciones basicas definidas de manera nativa en Python, es posible definir funciones nuevas que utilicen parametros e instrucciones especificas.

```python
# La estructura para construir una funcion es
# def nombre_funcion(parametros_funcion):
#   instrucciones 
```

Al igual que los ciclos, el bloque de instrucciones de la funcion debe ir indentado con respecto al codigo anterior.

```python
# Una funcion basica, sin parametros.
def fun1():
  print('hola')
```

```python
# La funcion queda guardada y debe ser llamada para ejecutarse
fun1()
```

```python
# Una funcion puede llevar parametros, con valores predefinidos
def fun2(a=3):
  print(a**2)
```

```python
fun2()
```

```python
# Se puede cambiar el valor del parametro para una ejecucion de la funcion
fun2(2)
```

```python
# Una funcion con muchos parametros los recibe uno a uno, 
# en el orden que fueron definidos
def fun3(a,b):
  print(a**b)
fun3(4,3)
```

Dentro de la funcion se pueden poner llamados a otras funciones, ciclos, condicionales y todo tipo de estructuras disponibles. Tambien es posible que la funcion regrese un valor, que se determina por medio de la instruccion return

```python
def fun4(a):
  return 3*a
```

```python
v=fun4(2)
v
```

La instruccion return tambien permite detener la ejecucion de la funcion a partir de una condicion. Existen muchas funciones nativas de Python que le permiten extraer informacion de los objetos o transformarlos de maneras sencillas, por ejemplo
* La funcion len() permite calcular la longitud de un objeto
* Para una cadena, los metodos .upper(), .upper(), .split() la convierten en minusculas, mayusculas o la dividen por espacios. A diferencia de una funcion, el metodo se aplica despues de la variable
* Para un diccionario, .keys(), .items() devuelven las claves y objetos.

```python
print(d.keys())
print(len(lista))
print('hola'.upper())
```
